```python
#import regular expression module in python
import re
```

```python
def wordopt(text):
  #convert to lowercase
  text = text.lower()
  #remove URLs
  text = re.sub(r'https?://\S+|www\.\S+', '', text)
  #remove HTML tags
  text = re.sub(r'<.*?>', '', text)
  #remove punctuation
  text = re.sub(r'[^\w\s]', '', text)
  #remove digits
  text = re.sub(r'\d', '', text)
  #remove newline characters
  text= re.sub(r'\n',' ',text)
  return text
```

```python
#TF vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

# Create and fit the vectorizer on the entire dataset (x)
vectorization = TfidfVectorizer()
xv = vectorization.fit_transform(x)

# Now split the data after vectorization
xv_train, xv_test, y_train, y_test = train_test_split(xv, y, test_size=0.3)

# Continue with the model training
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
#LR.fit(xv_train, y_train)

# --- SMOTE Integration Starts Here ---
!pip install imblearn
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
xv_train_resampled, y_train_resampled = smote.fit_resample(xv_train, y_train)
# --- SMOTE Integration Ends Here ---

LR.fit(xv_train_resampled, y_train_resampled)
```

```python
#prediction
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
# Import the necessary function
from sklearn.metrics import classification_report
#classification report
print(classification_report(y_test, pred_lr))
```
